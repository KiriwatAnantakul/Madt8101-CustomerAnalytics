## Word tokenization and dictionary

```python
stopwords = list(pythainlp.corpus.thai_stopwords())
removed_words = ['',' ','  ','\n', '(', ')','-','=',' ','/','.']
screening_words = stopwords + removed_words
```

```python
def tokenize_with_space(sentence):
  merged = ''
  words = pythainlp.word_tokenize(str(sentence), engine='newmm')
  for word in words:
    if word not in screening_words:
      merged = merged + ',' + word
  return merged[1:]
```

```python
df['Review_token'] = df['Review'].apply(lambda x: tokenize_with_space(x))
```

```python
df.tail()
```

```python
documents = df['Review_token'].to_list()
texts = [[text for text in doc.split(',')] for doc in documents]
dictionary = gensim.corpora.Dictionary(texts)
print(dictionary.token2id.keys())
```

```python
gensim_corpus = [dictionary.doc2bow(text, allow_update=True) for text in texts]
word_frequencies = [[(dictionary[id], frequence) for id, frequence in couple] for couple in gensim_corpus]
```

##Topic Modeling

```python
num_topics = 5
chunksize = 4000
passes = 20
iterations = 50
eval_every = 1

temp = dictionary[0]
id2word = dictionary.id2token

model = gensim.models.LdaModel(corpus=gensim_corpus, id2word=id2word, chunksize=chunksize, \
                              alpha='auto', eta='auto', \
                              iterations=iterations, num_topics=num_topics, \
                              passes=passes, eval_every=eval_every)
```

```python
vis = pyLDAvis.gensim.prepare(model, gensim_corpus, dictionary, mds='mmds')
vis
```

```python
from tabulate import tabulate

# Initialize a list to store the top topics for all 5 groups
top_topics_per_group = []

# Iterate through the 5 groups
for group_number in range(5):
    # Get the top topics for the current group
    top_topics = model.show_topic(group_number)

    # Append the top topics to the list with a label for the group
    top_topics_per_group.append({
        'Group': f'Group {group_number}',
        'Top Topics': ', '.join([f"{topic[0]} ({topic[1]:.3f})" for topic in top_topics])
    })

# Convert the list of dictionaries to a table and print it
table = tabulate(top_topics_per_group, headers='keys', tablefmt='pretty')
print(table)
```
